```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By
import time
```

```python
driver=webdriver.Chrome(r"D:\chromedriver.exe")
```

```python
driver.get("https://www.naukri.com/")
```

```python
designation=driver.find_element(By.CLASS_NAME,"suggestor-input ")
designation.send_keys('Data Analyst')
```

```python
location=driver.find_element(By.XPATH,"/html/body/div[1]/div[2]/div[3]/div/div/div[5]/div/div/div/input")
location.send_keys('Bangalore')
```

```python
job_title=[]
job_location=[]
company_name=[]
experience_required=[]
```

```python
#scraping job title from the given page
title_tags=driver.find_elements(By.XPATH,'//a[@class="title fw500 ellipsis"]')
for i in title_tags[0:10]:
    title=i.text
    job_title.append(title)
#scraping job location from the given page
location_tags=driver.find_elements(By.XPATH,'//li[@class="fleft grey-text br2 placeHolderLi location"]')
for i in location_tags[0:10]:
    location=i.text
    job_location.append(location)
#scraping company name from the given page
company_tags=driver.find_elements(By.XPATH,'//a[@class="subTitle ellipsis fleft"]')
for i in company_tags[0:10]:
    company=i.text
    company_name.append(company)
#scraping job experience from the given page
experience_tags=driver.find_elements(By.XPATH,'//li[@class="fleft grey-text br2 placeHolderLi experience"]//span')
for i in experience_tags[0:10]:
    exp=i.text
    experience_required.append(exp)
```

```python
print(len(job_title),len(job_location),len(company_name),len(experience_required))
```

```python
#creating the dataframe from above data-
df=pd.DataFrame({'job_title':job_title,'job_location':job_location,'company_name':company_name,'Exp_required':experience_required})
df
```

```python
company_name=[]
total_salary_record=[]
average_salary=[]
minimum_salary=[]
maximum_salary=[]
experience_required=[]
```

```python
#scraping company name from the given page
company_tags=driver.find_elements(By.XPATH,'/html/body/div/div/div/main/section[1]/div[2]/div[3]/div[2]/div[1]/div[1]/div/div/a')
for i in company_tags[0:10]:
    company=i.text
    company_name.append(company)
#scraping the total salary record from the given page
salary_tags=driver.find_elements(By.XPATH,'/html/body/div/div/div/main/section[1]/div[2]/div[3]/div[2]/div[1]/div[1]/div/div/div/span')
for i in salary_tags[0:10]:
    salary=i.text
    total_salary_record.append(salary)
#scraping the average salary from the given page
average_tags=driver.find_elements(By.XPATH,'/html/body/div/div/div/main/section[1]/div[2]/div[3]/div[2]/div[1]/div[2]/div/div[1]/div/p')
for i in average_tags[0:10]:
    average=i.text
    average_salary.append(average)
#scraping the minimum salary from the given page
minimum_tags=driver.find_elements(By.XPATH,'/html/body/div/div/div/main/section[1]/div[2]/div[3]/div[2]/div[1]/div[2]/div/div[2]/div[1]')
for i in minimum_tags[0:10]:
    minimum=i.text
    minimum_salary.append(minimum)
#scraping the maximim salary from the given data
maximum_tags=driver.find_elements(By.XPATH,'/html/body/div/div/div/main/section[1]/div[2]/div[3]/div[2]/div[1]/div[2]/div/div[2]/div[2]')
for i in maximum_tags[0:10]:
    maximum=i.text
    maximum_salary.append(maximum)
#scraping the experienced required from the given page
experience_tags=driver.find_elements(By.XPATH,'/html/body/div/div/div/main/section[1]/div[2]/div[3]/div[2]/div[1]/div[1]/div/div/div')
for i in experience_tags[0:10]:
    exp=i.text
    experience_required.append(exp)
```

```python
print(len(company_name),len(total_salary_record),len(average_salary),len(minimum_salary),len(maximum_salary),len(experience_required))
```
